```python
#Task 1
import numpy as np

arr = np.random.randint(0, 51, 20)
print("Масив:", arr)

threshold = int(input("Введи число: "))

count = np.sum(arr > threshold)
print("Кількість елементів більше:", count)

max_val = np.max(arr)
max_index = np.argmax(arr)
print("Максимум:", max_val)
print("Позиція:", max_index)

sorted_arr = np.sort(arr)[::-1]
print("Відсортований:", sorted_arr)

#Task 2
low = int(input("Нижня межа: "))
high = int(input("Верхня межа: "))

matrix = np.random.randint(low, high + 1, (5, 5))
print("Матриця:\n", matrix)

diag = np.diag(matrix)
print("Діагональ:", diag)

diag_sum = np.sum(diag)
print("Сума діагоналі:", diag_sum)

matrix[np.triu_indices(5, 1)] = 0
print("Після обнулення:\n", matrix)

#Task 3
start = int(input("Початок: "))
end = int(input("Кінець: "))

arr = np.arange(start, end + 1)

matrix = arr[:30].reshape(6, 5)  # 6x5 = 30
print("Матриця:\n", matrix)

row_sums = np.sum(matrix, axis=1)
print("Суми рядків:", row_sums)

col_max = np.max(matrix, axis=0)
print("Максимуми стовпців:", col_max)

#Task 4
low = int(input("Нижня межа: "))
high = int(input("Верхня межа: "))

arr = np.random.randint(low, high + 1, 15)
print("Масив:", arr)

negatives = arr[arr < 0]
print("Від'ємні:", negatives)

modified = arr.copy()
modified[modified < 0] = 0
print("Змінений:", modified)

zeros_count = np.sum(modified == 0)
print("Кількість нулів:", zeros_count)

#Task 5
n = int(input("Довжина масивів: "))

arr1 = np.random.randint(0, 11, n)
arr2 = np.random.randint(10, 21, n)

print("Масив 1:", arr1)
print("Масив 2:", arr2)

combined = np.concatenate((arr1, arr2))
print("Об'єднаний:", combined)

sum_arr = arr1 + arr2
diff_arr = arr1 - arr2

print("Сума:", sum_arr)
print("Різниця:", diff_arr)

#Task 6
rows = int(input("К-сть рядків: "))
cols = int(input("К-сть стовпців: "))

matrix = np.arange(1, rows * cols + 1).reshape(rows, cols)
print("Матриця:\n", matrix)

new_rows = int(input("Нові рядки: "))
new_cols = int(input("Нові стовпці: "))

new_matrix = matrix.reshape(new_rows, new_cols)
print("Нова матриця:\n", new_matrix)

row_min = np.min(new_matrix, axis=1)
row_max = np.max(new_matrix, axis=1)

print("Мін по рядках:", row_min)
print("Макс по рядках:", row_max)

total_sum = np.sum(new_matrix)
print("Сума всіх елементів:", total_sum)
```
