```python
#Task 1
import pandas as pd

data = {
    "Назва": ["Книга1", "Книга2", "Книга3", "Книга4", "Книга5", "Книга6"],
    "Автор": ["Автор1", "Автор2", "Автор3", "Автор4", "Автор5", "Автор6"],
    "Рік видання": [2010, 2018, 2020, 2015, 2017, 2022],
    "Ціна": [200, 350, 500, 150, 400, 600]
}

df = pd.DataFrame(data)

print(df)

print("Середня ціна:", df["Ціна"].mean())

print("Після 2015 року:\n", df[df["Рік видання"] > 2015])

print("Сортування за ціною:\n", df.sort_values(by="Ціна"))

#Task 2

df = pd.read_csv("orders.csv")

print(df.head(10))

print("Кількість замовлень:\n", df["Клієнт"].value_counts())

print("Макс:", df["Сума"].max())
print("Мін:", df["Сума"].min())

print("Загальна сума:", df["Сума"].sum())

#Task 3
data = {
    "Продукт": ["Хліб", "М'ясо", "Сир", "Яблуко", "Банан", "Рис", "Курка", "Риба", "Шоколад", "Молоко"],
    "Категорія": ["Зернові", "М'ясо", "Молочні", "Фрукти", "Фрукти", "Зернові", "М'ясо", "М'ясо", "Солодощі", "Молочні"],
    "Калорії": [250, 500, 350, 80, 90, 300, 400, 350, 550, 120],
    "Білки": [8, 25, 20, 1, 1, 7, 30, 22, 5, 6]
}

df = pd.DataFrame(data)

print(df)

print("Калорії > 300:\n", df[df["Калорії"] > 300])

print("Середній білок:\n", df.groupby("Категорія")["Білки"].mean())

print("Сортування:\n", df.sort_values(by="Калорії", ascending=False))

#Task 4
data = {
    "Ім'я": ["Іван", "Оля", "Петро", "Іван", "Оля", "Петро", "Анна", "Анна"],
    "Проект": ["A", "A", "B", "B", "C", "C", "A", "B"],
    "Години": [10, 15, 20, 25, 30, 10, 5, 35]
}

df = pd.DataFrame(data)

print(df)

print("Години по співробітниках:\n", df.groupby("Ім'я")["Години"].sum())

print("Години по проектах:\n", df.groupby("Проект")["Години"].sum())

total_hours = df.groupby("Ім'я")["Години"].sum()
print("Найбільше годин:", total_hours.idxmax())

# Task 5
data = {
    "Фільм": ["A","A","A","B","B","B","C","C","C","D","D","D"],
    "Місто": ["Київ","Львів","Одеса","Київ","Львів","Одеса","Київ","Львів","Одеса","Київ","Львів","Одеса"],
    "Продані квитки": [100,80,60,120,90,70,150,110,100,200,180,160]
}

df = pd.DataFrame(data)

print(df)

print("По фільмах:\n", df.groupby("Фільм")["Продані квитки"].sum())

print("По містах:\n", df.groupby("Місто")["Продані квитки"].sum())

film_sales = df.groupby("Фільм")["Продані квитки"].sum()
print("Найпопулярніший фільм:", film_sales.idxmax())

#Task 6 
clients = pd.DataFrame({
    "ID": [1,2,3,4,5],
    "Ім'я": ["Іван", "Оля", "Петро", "Анна", "Макс"],
    "Місто": ["Київ", "Львів", "Одеса", "Харків", "Дніпро"]
})

orders = pd.DataFrame({
    "ID клієнта": [1,2,1,3,5],
    "Замовлення": ["A","B","C","D","E"],
    "Вартість": [1000,1500,500,700,1200]
})

print(clients)
print(orders)

merged = pd.merge(clients, orders, left_on="ID", right_on="ID клієнта")

result = merged.groupby("Ім'я")["Вартість"].sum().reset_index()

print("Результат:\n", result)
```
