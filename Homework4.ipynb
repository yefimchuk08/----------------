```python
import matplotlib.pyplot as plt
import numpy as np

#Task 1
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
plan = [100, 120, 130, 150, 170, 200]
fact = [90, 110, 140, 145, 180, 210]

plt.plot(months, plan, label="План")
plt.plot(months, fact, label="Факт")

plt.xlabel("Місяці")
plt.ylabel("Продажі")
plt.title("План vs Факт")
plt.legend()

plt.show()

#Task 2

ages = np.random.randint(18, 65, 100)

mean_age = np.mean(ages)

plt.hist(ages)

plt.axvline(mean_age)

plt.xlabel("Вік")
plt.ylabel("Кількість")
plt.title("Розподіл віку")

plt.show()

#Task 3


group1 = np.random.randint(60, 100, 20)
group2 = np.random.randint(50, 95, 20)
group3 = np.random.randint(70, 100, 20)

plt.boxplot([group1, group2, group3])

plt.xticks([1, 2, 3], ["Група 1", "Група 2", "Група 3"])

plt.xlabel("Групи")
plt.ylabel("Оцінки")
plt.title("Порівняння результатів")

plt.show()

#Task 4

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temp = [20, 22, 19, 23, 25, 24, 21]
humidity = [60, 65, 55, 70, 75, 80, 68]

plt.plot(days, temp, label="Температура")
plt.plot(days, humidity, label="Вологість")

plt.xlabel("Дні")
plt.ylabel("Значення")
plt.title("Погода за тиждень")
plt.legend()

plt.xticks(rotation=45)

plt.show()

#Task 5

hours = list(range(24))
load = [10, 12, 8, 6, 5, 7, 15, 30, 50, 60, 70, 65, 60, 55, 50, 45, 40, 35, 30, 25, 20, 18, 15, 12]

plt.plot(hours, load)

plt.fill_between(hours, load)

plt.grid()

plt.xlabel("Години")
plt.ylabel("Навантаження")
plt.title("Навантаження сервера")

plt.show()

#Task 6
time = list(range(10))

conversion = np.random.rand(10)
retention = np.random.rand(10)
avg_check = np.random.randint(100, 500, 10)
orders = np.random.randint(10, 100, 10)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(time, conversion)
axs[0, 0].set_title("Конверсія")

axs[0, 1].plot(time, retention)
axs[0, 1].set_title("Утримання")

axs[1, 0].plot(time, avg_check)
axs[1, 0].set_title("Середній чек")

axs[1, 1].plot(time, orders)
axs[1, 1].set_title("Замовлення")

plt.suptitle("Метрики продукту")

plt.show()
```
