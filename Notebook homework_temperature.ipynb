```python
# Task 1
temperatures = [2, 5, 4, 6, 7, 3, 1]
average_temp = sum(temperatures) / len(temperatures)
max_temp = max(temperatures)
min_temp = min(temperatures)
difference = max_temp - min_temp
average_temp, max_temp, min_temp, difference

# Task 2

expenses = [1200, 1500, 800, 600, 1000, 400, 900]

total = sum(expenses)
avg = total / len(expenses)
mx = max(expenses)
mn = min(expenses)
days_over_1000 = sum(1 for x in expenses if x > 1000)

total, avg, mx, mn, days_over_1000

# Task 3

votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]

from collections import Counter

counter = Counter(votes)

leader = counter.most_common(1)[0]
top3 = counter.most_common(3)

counter, leader, top3

# Task 4

math = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]

def stats(arr):
    return sum(arr)/len(arr), min(arr), max(arr)

math_stats = stats(math)
physics_stats = stats(physics)
english_stats = stats(english)

subjects = {
    "math": math_stats[0],
    "physics": physics_stats[0],
    "english": english_stats[0]
}

best_subject = max(subjects, key=subjects.get)

math_stats, physics_stats, english_stats, best_subject
```
